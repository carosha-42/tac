```python
# Instantier l'extracteur de mots clés
kw_extractor = yake.KeywordExtractor(lan="fr", top=50)
kw_extractor
```

```python
# Définir le chemin des données
txt_path = "C:/Users/duboi/MA2TAC/tac/data/txt"
```

```python
# Choisir l'année souhaitée
year = "1941"

# Lister les fichiers de l'année choisie (1939)
files = [f for f in os.listdir(txt_path) if f.endswith('.txt') and year in f]
```

```python
# Imprimer le nombre de fichiers identifiés pour l'année 1939
len(files)
```

```python
# Extraction des mots-clés pour chaque fichier de l'année choisie
for f in sorted(files)[:10]:  
    # Lecture du texte du fichier
    file_path = os.path.join(txt_path, f)
    with open(file_path, 'r', encoding="utf-8") as file:
        text = file.read()
```

```python
# Afficher le contenu du fichier pour vérification
print(f"Contenu de {f} :\n{text[:500]}\n")
```

```python
# Extraire les mots-clés
keywords = kw_extractor.extract_keywords(text)
keywords
```

```python
# Ne garder que les bigrammes
kept = []
for kw, score in keywords:
    words = kw.split()
    if len(words) == 2:
        kept.append(kw)
kept
```

```python
# Afficher les mots-clés extraits
print(f"{f} mentions ces mots-clés : {', '.join(kept)}...")
```

```python
#Import des stopwords 
from collections import Counter
from wordcloud import WordCloud
import os
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from IPython.display import Image
```

```python
# Stopwords 
sw = stopwords.words("french")
sw += ["les", "plus", "cette", "fait", "faire", "être", "deux", "comme", "dont", "tout",
       "ils", "bien", "sans", "peut", "tous", "après", "ainsi", "donc", "cet", "sous",
       "celle", "entre", "encore", "toutes", "pendant", "moins", "dire", "cela", "non",
       "faut", "trois", "aussi", "dit", "avoir", "doit", "contre", "depuis", "autres",
       "van", "het", "autre", "jusqu", "ville", "rossel", "dem"]
sw = set(sw)
```

```python
# Stocker le contenu de ces fichiers dans une liste
content_list = []
for txt in txts:
    with open(os.path.join(txt_path, txt), 'r', encoding='utf-8') as f:
        content_list.append(f.read())
```

```python
# Ecrire tout le contenu dans un fichier temporaire
temp_path = 'C:/Users/duboi/MA2TAC/tac/data/tmp'
if not os.path.exists(temp_path):
    os.mkdir(temp_path)
with open(os.path.join(temp_path, f'{year}.txt'), 'w', encoding='utf-8') as f:
    f.write(' '.join(content_list))
```

```python
# Imprimer le contenu du fichier et constater les "déchets"
with open(os.path.join(temp_path, f'{year}.txt'), 'r', encoding='utf-8') as f:
    before = f.read()

before[:500]
```

```python
#Création fonction nettoyage
def clean_text(year, folder=None):
    if folder is None:
        input_path = f"{year}.txt"
        output_path = f"{year}_clean.txt"
    else:
        input_path = f"{folder}/{year}.txt"
        output_path = f"{folder}/{year}_clean.txt"
    output = open(output_path, "w", encoding='utf-8')
    with open(input_path, 'r', encoding='utf-8') as f:
        text = f.read()
        words = nltk.wordpunct_tokenize(text)
        kept = [w.upper() for w in words if len(w) > 2 and w.isalpha() and w.lower() not in sw]
        kept_string = " ".join(kept)
        output.write(kept_string)
    return f'Output has been written in {output_path}!'
```

```python
#Application de la fonction
clean_text(year, folder=temp_path)
```

```python
# Vérifier le résultat
with open(os.path.join(temp_path, f'{year}_clean.txt'), 'r', encoding='utf-8') as f:
    after = f.read()

after[:500]
```

```python
#Affichage des mots les plus fréquents
frequencies = Counter(after.split())
print(frequencies.most_common(10))
```

```python
#Nuage de mots!
cloud = WordCloud(width=2000, height=1000, background_color='white').generate_from_frequencies(frequencies)
cloud.to_file(os.path.join(temp_path, f"{year}.png"))
Image(filename=os.path.join(temp_path, f"{year}.png"))
```

```python
from collections import defaultdict
import os
import spacy

# Charger le modèle de langue français de spaCy
!python -m spacy download fr_core_news_md
nlp = spacy.load('fr_core_news_md')
```

```python
# Définir le chemin vers le dossier contenant les fichiers texte et l'année cible
txt_path = "C:/Users/duboi/MA2TAC/tac/data/txt"
year = "1941"

# Filtrer les fichiers de l'année 1941
files = [f for f in os.listdir(txt_path) if f.endswith('.txt') and year in f]

# Charger le contenu des fichiers de l'année 1941
text = ""
for f in files:
    with open(os.path.join(txt_path, f), 'r', encoding='utf-8') as file:
        text += file.read() + " "  # Ajouter un espace entre chaque fichier pour éviter de coller les textes

# Limiter la taille du texte si nécessaire (pour gérer les grands volumes)
n = 1000000  # Limite de caractères à traiter
text = text[:n]
```

```python
# Traiter le texte avec spaCy pour extraire les entités nommées
doc = nlp(text)

# Compter les entités nommées
people = defaultdict(int)
organizations = defaultdict(int)
locations = defaultdict(int)

for ent in doc.ents:
    if ent.label_ == "PER" and len(ent.text) > 3:  # Personnes
        people[ent.text] += 1
    elif ent.label_ == "ORG" and len(ent.text) > 3:  # Organisations
        organizations[ent.text] += 1
    elif ent.label_ == "LOC" and len(ent.text) > 3:  # Lieux
        locations[ent.text] += 1
```

```python
# Trier les entités nommées par fréquence et imprimer les principales pour chaque catégorie
print("\nPersonnes les plus fréquentes :")
sorted_people = sorted(people.items(), key=lambda kv: kv[1], reverse=True)
for person, freq in sorted_people[:50]:
    print(f"{person} apparait {freq} fois")
```

```python
print("\nOrganisations les plus fréquentes :")
sorted_organizations = sorted(organizations.items(), key=lambda kv: kv[1], reverse=True)
for org, freq in sorted_organizations[:50]:
    print(f"{org} apparait {freq} fois")
```

```python
print("\nLieux les plus fréquents :")
sorted_locations = sorted(locations.items(), key=lambda kv: kv[1], reverse=True)
for loc, freq in sorted_locations[:50]:
    print(f"{loc} apparait {freq} fois")
```
